```python
import os
import os.path
```

```python
a = []
for root, dirs, files in os.walk("/Users/ciwwwnd/Desktop/тест/", topdown=False):
    for name in files:
        a.append(os.path.join(root, name))

links = []
for url in a:
    if 'message' in url:
        links.append(url)
        
```

```python
from bs4 import BeautifulSoup
import re
import string
from pymystem3 import Mystem
m = Mystem()
```

```python
def lemmatize(text):
    lemmatised_sents = []
    split_r = re.compile(r'[.|!|?|…|[A-Za-z]|[\d]+]')
    sentences = filter(lambda t: t, [t.strip() for t in split_r.split(text)])
    for sentence in sentences:
        _ = str.maketrans(dict.fromkeys(string.punctuation))
        without_puncts = sentence.translate(_).lower()            
        lemmas = ''.join(m.lemmatize(without_puncts))
        lemmatised_sents.append(lemmas)
    lemmatised_text = '\n'.join(lemmatised_sents)
    return lemmatised_text
```

Записываем новые файлы, 1 файл lemmatised_messages соответствует одному файлу messages

```python
for link in links:
    comments = []
    with open(link, encoding='utf-8') as f:
        content = f.read()
        soup = BeautifulSoup(content, 'lxml')
        text = soup.find_all('div', class_= 'text')
        for t in text:
            cleantext = BeautifulSoup(str(t), 'lxml').text
            lemmatised = lemmatize(cleantext)
            comments.append(lemmatised)
                    
    filename = re.findall('(.*)messages', link)[0] + 'lemmatised_messages' + re.findall('messages(.*)\.html', link)[0] + '.txt'
    with open(filename, 'w', encoding = 'utf-8') as txtfile:
        for line in comments:
            txtfile.write(line)
```

Считаем слова. Была взята тестовая папка, в которой просто лежат только несколько файлов lemmatised_messages, так как весь корпус обрабатывался бы очень долго. 

```python
from collections import Counter
import re
from tqdm.auto import tqdm
import csv 
```

```python
words = []
for i in range(1, 2):
    with open('/Users/ciwwwnd/Desktop/test/lemmatised_messages' + str(i) + '.txt', 'r', encoding='utf-8') as message:
        mes = message.read().lower()
        words.extend(re.findall('[a-zа-яё]+', mes))
cnt = dict(Counter(words))
```

```python
cnt
```

# ipm

```python
ipm = {}
for word in list(cnt):
    ipm[word] = cnt[word]/len(words)*1000000
    
```

```python
ipm
```

```python
def find_indexes(num_of_segments):
    a = 0
    indexes = []
    for i in range(num_of_segments):
        indexes.append([a, a + round(len(words)/num_of_segments)])
        a += round(len(words)/num_of_segments)
    indexes[-1][1] = len(words)-1
    return indexes
```

```python
indexes = find_indexes(100)
```

```python
vs = {} # средняя частота по сегментам
for word in tqdm(range(len(list(cnt)))):
    v = 0
    for ind in indexes:
        v += len(re.findall(list(cnt)[word], ' '.join(words[ind[0]:ind[1]])))/len(words[ind[0]:ind[1]])
    vs[list(cnt)[word]] = v/100
    
```

```python
n = 100
sigmas = {}
for word in tqdm(range(len(list(cnt)))):
    sum_v = 0
    for ind in indexes:
        v = len(re.findall(list(cnt)[word], ' '.join(words[ind[0]:ind[1]])))/len(words[ind[0]:ind[1]])
        sum_v += (v - vs[list(cnt)[word]])**2
    sigmas[list(cnt)[word]] = (1/(n-1)*sum_v)**0.5
    
```

# Juilland's D

```python
d_juilland = {}
for word in tqdm(range(len(list(cnt)))):
    d_juilland[list(cnt)[word]] = 100*(1-(sigmas[list(cnt)[word]]/(vs[list(cnt)[word]]*(n**0.5))))
    
```

```python
d_juilland
```

# Juilland's U

```python
u_juilland = {}
for word in tqdm(range(len(list(cnt)))):
    u_juilland[list(cnt)[word]] = d_juilland[list(cnt)[word]]*cnt[list(cnt)[word]]
    
```

```python
u_juilland
```

```python
words_with_ind = {} 
for i in range(1, len(words)):
    try:
        words_with_ind[words[i]].append(i)
    except:
        words_with_ind[words[i]] = [i]
        
```

# ARF

```python
ARF = {}
for i in tqdm(range(len(words))):
    word = words[i]
    f = cnt[word]
    N = len(words)
    v = N/f
    ds = [float(v)]
    if len(words_with_ind[word]) == 1:
        ds.append(1)
    if len(words_with_ind[word]) > 1:
        ds.append(N-words_with_ind[word][-1]+words_with_ind[word][0])
        for i in range(1,len(words_with_ind[word])):
            ds.append(words_with_ind[word][i]-words_with_ind[word][i-1])
    ds.sort()
    ds = ds[:f]
    num = 0
    for j in ds:
        num+=j
    if f == 1:
        ARF[word] = 1.0
    else:
        ARF[word] = f/N*num
    
```

Записываем результаты в csv-файл. 

```python
ARF_l = []
ipm_l = []
u_juilland_l = []
d_juilland_l = []
ws = list(cnt)
for word in ws:
    ARF_l.append(ARF[word])
    ipm_l.append(ipm[word])
    u_juilland_l.append(u_juilland[word])
    d_juilland_l.append(d_juilland[word])
    
```

```python
with open('results_telegram.csv', 'a', encoding='utf-8') as csv_file:
    file_writer = csv.writer(csv_file, delimiter = "\t")
    file_writer.writerow(["Word", "ipm", "ARF", "Juilland's D", "Juilland's U"])
    for j in tqdm(range(len(ws))):
        file_writer.writerow([ws[j], ipm_l[j], ARF_l[j], d_juilland_l[j], u_juilland_l[j]])
```
